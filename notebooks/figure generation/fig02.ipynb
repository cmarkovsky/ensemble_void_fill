```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
plt.rcParams["mathtext.fontset"] = "dejavusans"
rc('mathtext', fontset='stixsans')
# rc('text.latex', preamble=r'\usepackage{cmbright}')
```

```python
filename = 'results_final/combined_final.feather'
combined = pd.read_feather(filename)
ehim = combined[combined['reg'] == 'ehim']
whim = combined[combined['reg'] == 'whim']
```

```python
## Calculating bins/statistics for E. Himalaya

ez1 = ehim[ehim['void_mask'] == False]['z']
edh1 = ehim[ehim['void_mask'] == False]['target']

ez2 = ehim[ehim['void_mask'] == True]['z']
edh2 = ehim[ehim['void_mask'] == True]['target']


ebin_edges = np.arange(min(ehim['z']), max(ehim['z']), 50)
ebin_indices = np.digitize(ez1, ebin_edges)  # Get bin indices for each elevation point
eaverage_dh1 = [
    np.mean(edh1[ebin_indices == i]) if len(edh1[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

emed_dh1 = [
    np.median(edh1[ebin_indices == i]) if len(edh1[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

estd_dh1 = [
    np.std(edh1[ebin_indices == i]) if len(edh1[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

ebin_indices = np.digitize(ez2, ebin_edges)  # Get bin indices for each elevation point
eaverage_dh2 = [
    np.mean(edh2[ebin_indices == i]) if len(edh2[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

emed_dh2 = [
    np.median(edh2[ebin_indices == i]) if len(edh2[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

estd_dh2 = [
    np.std(edh2[ebin_indices == i]) if len(edh2[ebin_indices == i]) > 0 else np.nan
    for i in range(1, len(ebin_edges))
]

# Calculating bins/statistics for W. Himalaya

wz1 = whim[whim['void_mask'] == False]['z']
wdh1 = whim[whim['void_mask'] == False]['target']

wz2 = whim[whim['void_mask'] == True]['z']
wdh2 = whim[whim['void_mask'] == True]['target']

wbin_edges = np.arange(min(whim['z']), max(whim['z']), 50)
wbin_indices = np.digitize(wz1, wbin_edges)  # Get bin indices for each elevation point
waverage_dh1 = [
    np.mean(wdh1[wbin_indices == i]) if len(wdh1[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]

wmed_dh1 = [
    np.median(wdh1[wbin_indices == i]) if len(wdh1[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]

wstd_dh1 = [
    np.std(wdh1[wbin_indices == i]) if len(wdh1[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]

wbin_indices = np.digitize(wz2, wbin_edges)  # Get bin indices for each elevation point
waverage_dh2 = [
    np.mean(wdh2[wbin_indices == i]) if len(wdh2[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]

wmed_dh2 = [
    np.median(wdh2[wbin_indices == i]) if len(wdh2[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]

wstd_dh2 = [
    np.std(wdh2[wbin_indices == i]) if len(wdh2[wbin_indices == i]) > 0 else np.nan
    for i in range(1, len(wbin_edges))
]
```

```python
plot_ehim = True

fig, axs = plt.subplot_mosaic([['a)', 'b)']], figsize = (20,10), sharex = True, sharey = True, layout = 'constrained')

for label, ax in axs.items():
    ax.set_title(label, fontfamily='sans-serif', loc='left', fontsize=16)
    
axes1 = axs['a)']
axes2 = axs['b)']

sns.set_style('white')

ax11 = sns.histplot(data = whim, x = 'z', hue = 'void_mask', kde = False, bins = wbin_edges, ax = axes1,
                   multiple = 'stack', hue_order = [False, True], legend = False, alpha = 0.6, palette = [sns.color_palette()[0], sns.color_palette()[1]])

ax12 = ax11.twinx()
ax12.plot(
    wbin_edges[:-1] + 25,
    waverage_dh2,
    marker='o',
    color = sns.color_palette()[1],
    markersize=4,
    linewidth=1,
    # label="Average dh",
)

ax12.plot(
    wbin_edges[:-1] + 25,
    waverage_dh1,
    marker='o',
    color = sns.color_palette()[0],
    markersize=4,
    linewidth=1,
    # label="Average dh",
)

ax12.fill_between(
    wbin_edges[:-1] + 25,
    np.array(waverage_dh2) - np.array(wstd_dh2),
    np.array(waverage_dh2) + np.array(wstd_dh2),
    color=sns.color_palette()[1],
    alpha=0.2,
    # label="±1 Std. Dev.",
)
ax12.fill_between(
    wbin_edges[:-1] + 25,
    np.array(waverage_dh1) - np.array(wstd_dh1),
    np.array(waverage_dh1) + np.array(wstd_dh1),
    color=sns.color_palette()[0],
    alpha=0.2,
    # label="±1 Std. Dev.",
)

ax11.grid(visible=False)  # Disable gridlines on secondary axis
ax12.grid(visible=False)  # Disable gridlines on secondary axis
if plot_ehim:
    ax21 = sns.histplot(data = ehim, x = 'z', hue = 'void_mask', kde = False, bins = ebin_edges, ax = axes2,
                       multiple = 'stack', hue_order = [False, True], legend = False, alpha = 0.6, palette = [sns.color_palette()[0], sns.color_palette()[1]])
    # ax1 = sns.histplot(data = full, x = 'z', hue = 'void_mask', kde = False, bins = bin_edges, 
    #                    multiple = 'stack', hue_order = [False, True], legend = False, alpha = 0.6, palette = ["#0072B2", "#E69F00"])
    ax22 = ax21.twinx()
    # color_temp = sns.color_palette("Reds", n_colors=9)[5]
    ax22.plot(
        ebin_edges[:-1] + 25,
        eaverage_dh2,
        marker='o',
        color = sns.color_palette()[1],
        markersize=4,
        linewidth=1,
        # label="Average dh",
    )
    
    ax22.plot(
        ebin_edges[:-1] + 25,
        eaverage_dh1,
        marker='o',
        color = sns.color_palette()[0],
        markersize=4,
        linewidth=1,
        # label="Average dh",
    )
    
    ax22.fill_between(
        ebin_edges[:-1] + 25,
        np.array(eaverage_dh2) - np.array(estd_dh2),
        np.array(eaverage_dh2) + np.array(estd_dh2),
        color=sns.color_palette()[1],
        alpha=0.2,
        # label="±1 Std. Dev.",
    )
    ax22.fill_between(
        ebin_edges[:-1] + 25,
        np.array(eaverage_dh1) - np.array(estd_dh1),
        np.array(eaverage_dh1) + np.array(estd_dh1),
        color=sns.color_palette()[0],
        alpha=0.2,
        # label="±1 Std. Dev.",
    )
# # Align the zero of both axes
ax11.set_ylim(-200000, 350000)
ax12.set_ylim(-2, 3.5)
ax22.set_ylim(-2, 3.5)
# Add legends
ax11.legend(loc="upper left", fontsize=16, frameon=True, labels = ['Artificial void', 'Non-void'])
ax21.legend(loc="upper left", fontsize=16, frameon=True, labels = ['Artificial void', 'Non-void'])
# ax2.legend(loc="upper right", fontsize=12, frameon=False)

ax11.axhline(0, color = 'grey', linestyle = '--', linewidth = 0.75)
ax21.axhline(0, color = 'grey', linestyle = '--', linewidth = 0.75)

ax11.set_ylabel("Number of pixels", fontsize=16, labelpad=10)
ax21.set_ylabel("", fontsize=12, labelpad=10)
ax12.set_ylabel("", fontsize=12, labelpad=5)
axes1.set_xticks([3000, 4000, 5000, 6000, 7000], 
                 labels = ['3000', '4000', '5000', '6000', '7000'])
ax11.set_yticks([0, 100000, 200000, 300000], labels = ['0', '100000', '200000', '300000'])
ax22.set_yticks([-2, -1, 0, 1, 2, 3], labels = ['-2', '-1', '0', '1', '2', '3'])
ax12.set_yticks([])
ax11.set_xlabel("Elevation (m)", fontsize=16, labelpad=10)
ax21.set_xlabel("Elevation (m)", fontsize=16, labelpad=10)
ax22.set_ylabel("Surface elevation change (m a$^{-1}$)", fontsize=16, labelpad=20, rotation = -90)
ax11.tick_params(axis='both', which='major', labelsize=16)
ax21.tick_params(axis='both', which='major', labelsize=16)
ax22.tick_params(axis='both', which='major', labelsize=16)
axes1.set_title('Western Himalaya', fontsize = 16)
axes2.set_title('Eastern Himalaya', fontsize = 16)
# fig.tight_layout()
# plt.show()
plt.savefig('./figures/bothreg_hist_FINAL.png', bbox_inches='tight', dpi = 300)
```

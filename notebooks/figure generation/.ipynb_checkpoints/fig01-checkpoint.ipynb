```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.image as mpimg
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# rc('font',**{'family':'serif','serif':['Times']})
rc('text', usetex=True)
plt.rcParams["mathtext.fontset"] = "dejavusans"
rc('mathtext', fontset='stixsans')
```

```python
filename = './results_final/combined_final.feather'
results = pd.read_feather(filename)

mb_data = pd.read_csv('../NEW_VOID/dshean/data/mb/hma_mb_20190215_0815_rmse.csv')
```

```python
ehim = results[results['reg'] == 'ehim']
whim = results[results['reg'] == 'whim']
```

```python
mb_ehim = mb_data[mb_data['RGIId'].isin(ehim['RGIId'].unique())]
mb_ehim['reg'] = 'ehim'
mb_whim = mb_data[mb_data['RGIId'].isin(whim['RGIId'].unique())]
mb_whim['reg'] = 'whim'
```

```python
ehim_sort = mb_ehim.sort_values(by='dhdt_ma', ascending = True)
# ehim_sort[['RGIId', 'dhdt_ma', 'area_m2']]
whim_sort = mb_whim.sort_values(by='dhdt_ma', ascending = False)
```

```python
mb_combined = pd.concat([mb_ehim, mb_whim])
```

```python
sns.set_style('white')
def plot_map_stats(mb):
    fig, axes = plt.subplot_mosaic([['a)', 'b)'], ['a)', 'c)']], figsize = (20,10), width_ratios = [2.5, 1.5], sharex = False, sharey = False, layout = 'constrained')
    for label, ax in axes.items():
        # if label == 'a)':
        #     label = '    '.join(label)
        ax.set_title(label, fontfamily='sans-serif', loc='left', fontsize=16)
    ax0 = axes['a)']
    ax1 = axes['b)']
    ax2 = axes['c)']
    # hist1 = sns.histplot(data = mb_combined, x = 'z_med', hue = 'reg', hue_order = ['ehim', 'whim'], element = 'step', ax = ax1, palette = ['#FF0000', '#00FFFF'], legend = False)
    hist1 = sns.histplot(data = mb, x = 'z_med', hue = 'reg', ax = ax1, hue_order = ['ehim', 'whim'], palette = [sns.color_palette('Set1')[0], sns.color_palette('Set1')[1]], binwidth = 50, alpha = 0.8, legend = False)
    hist2 = sns.histplot(data = mb, x = 'dhdt_ma', hue = 'reg', ax = ax2, hue_order = ['ehim', 'whim'], palette = [sns.color_palette('Set1')[0], sns.color_palette('Set1')[1]], alpha = 0.8, legend = False)
    ax2.set_xlim([-2, 2])
    
    ax1.set_xlabel('Median Elevation (m)', fontsize = 16, labelpad = 10)
    ax2.set_xlabel('Volume change (m a$^{-1}$)', fontsize = 16, labelpad = 10)

    ax1.set_ylabel('Number of glaciers', fontsize = 16, labelpad = 10)
    ax2.set_ylabel('Number of glaciers', fontsize = 16, labelpad = 10)

    ax1.set_xticks([3500, 4000, 4500, 5000, 5500, 6000, 6500], labels = ['3500', '4000', '4500', '5000','5500', '6000', '6500'])
    ax1.set_yticks([0, 100, 200, 300, 400], labels = ['0', '100', '200', '300', '400'])
    ax1.tick_params(axis='both', which='major', labelsize=16)
    
    ax2.set_xticks([-2, -1.5, -1, -0.5, 0, 0.5, 1, 1.5, 2], labels = ['-2', '-1.5', '-1', '-0.5', '0', '0.5', '1', '1.5', '2'])
    ax2.set_yticks([0, 100, 200, 300, 400, 500], labels = ['0', '100', '200', '300', '400', '500'])
    ax2.tick_params(axis='both', which='major', labelsize=16)

    whim_patch = mpatches.Patch(color=sns.color_palette('Set1')[1], label='Western Himalaya')
    ehim_patch = mpatches.Patch(color=sns.color_palette('Set1')[0], label='Eastern Himalaya')
    handles = [whim_patch, ehim_patch]
    ax1.legend(handles = handles, loc = 'upper left', prop={'size': 16})
    ax2.legend(handles = handles, loc = 'upper left', prop={'size': 16})
    
    ax0.axis('off')
    try:
        img = mpimg.imread('./figures/reg_map_FINAL.png')
    except FileNotFoundError:
        print("Image file not found. Please replace 'your_image.png' with a valid path to an image.")
        # Create a dummy image for demonstration if the file is not found
        img = np.random.rand(100, 100, 3)
    ax0.imshow(img)
    ax0.set_ymargin(-0.04)
    plt.savefig('./figures/reg_stats_FINAL.png', bbox_inches='tight', dpi = 300)
    
plot_map_stats(mb_combined)
```
